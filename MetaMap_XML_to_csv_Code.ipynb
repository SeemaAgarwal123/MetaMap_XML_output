```python
# for abbreviated text in xml MM output

#for testing
#tree=et.parse(r'218-01.out')
#root=tree.getroot()
def dfabbreviation():
    AAsCount=[]
    AAText=[]
    AAExp=[]
    AATokenNum=[]
    AALen=[]
    AAExpTokenNum=[]
    AAExpLen=[]
    AACUIsCount=[]
    AACUI=[]

    for x in root.iter('AAs'):
        root1=et.Element('root')
        root1=x
        cnt=x.attrib['Count']
        if (cnt!='0'):
            AAsCount.append(cnt)
            for AA in root1.iter('AA'):
                root2=et.Element('root')
                root2=AA    
                for t in root2.iter('AAText'):
                    AAText.append(t.text)
                for ex in root2.iter('AAExp'):
                    AAExp.append(ex.text)
                for tnum in root2.iter('AATokenNum'):
                    AATokenNum.append(tnum.text)
                for aalen in root2.iter('AALen'):
                    AALen.append(aalen.text)
                for exptnum in root2.iter('AAExpTokenNum'):
                    AAExpTokenNum.append(exptnum.text)
                for explen in root2.iter('AAExpLen'):
                    AAExpLen.append(explen.text)

                for supply in root1.iter('AACUIs'):
                    root2=et.Element('root')
                    root2=(supply)
                    AACUIsCount.append(supply.attrib['Count'])
                    if (cnt!='0'):
                        for tech in root2.iter('AACUI'):
                            AACUI.append(tech.text)


    dict_AA = {
        'AAsCount':AAsCount,
    'AAText':AAText,
    'AAExp':AAExp,
    'AATokenNum':AATokenNum,
    'AALen':AALen,
    'AAExpTokenNum':AAExpTokenNum,
    'AAExpLen':AAExpLen,
    'AACUIsCount':AACUIsCount,
    'AACUI':AACUI
    }


    df_AA=pd.DataFrame(dict([(col_name,pd.Series(values)) for col_name,values in dict_AA.items() ]))
    return df_AA            

```

```python
# for negated text in xml MM output


#for testing
#tree=et.parse(r'132-05.out')
#root=tree.getroot()
def dfnegation():
    NegCount=[]
    NegType=[]
    NegTrigger=[]
    NegTriggerPIsCount=[]
    NegTriggerPIStart=[]
    NegTriggerPILength=[]
    NegConceptsCount=[]
    NegConceptCUI=[]
    NegConceptMatched=[]
    NegConcPIsCount=[]
    NegConcPIStart=[]
    NegTriggerPIStart=[]
    NegConcPILength=[]

    

    for x in root.iter('Negations'):
        root1=et.Element('root')
        root1=x
        cnt=x.attrib['Count']
        if (cnt!='0'):
            NegCount.append(cnt)
            for supply in root1.iter('Negation'):
                root2=et.Element('root')
                root2=(supply)
                for tech in root2.iter('NegType'):
                    NegType.append(tech.text)
                for yr in root2.iter('NegTrigger'):
                    NegTrigger.append(yr.text)            
                for gas in root2.iter('NegTriggerPIs'):
                    root3 = et.Element('root')
                    root3=gas
                    gas.attrib['Count']
                    NegTriggerPIsCount.append(gas.attrib['Count'])
                    for em in root3.iter('NegTriggerPI'):
                        root3a=et.Element('root')
                        root3a=em
                        for strtpos in root3a.iter('StartPos'):
                            NegTriggerPIStart.append(strtpos.text)
                        for leng in root3a.iter('Length'):
                            NegTriggerPILength.append(leng.text)                

                for concept in root2.iter('NegConcepts'):
                    root4 = et.Element('root')
                    root4=concept
                    NegConceptsCount.append(concept.attrib['Count'])
                    for c in root4.iter('NegConcept'):
                        root4a=et.Element('root')
                        root4a=c                
                        for concui in root4a.iter('NegConcCUI'):
                            NegConceptCUI.append(concui.text)
                        for conMatch in root4a.iter('NegConcMatched'):
                            NegConceptMatched.append(conMatch.text)
                for conceptPIs in root2.iter('NegConcPIs'):
                    root5 = et.Element('root')
                    root5=conceptPIs
                    NegConcPIsCount.append(conceptPIs.attrib['Count'])
                    for concpi in root5.iter('NegConcPI'):
                        root5a=et.Element('root')
                        root5a=concpi 

                        for NegConcstartPos in root5a.iter('StartPos'):
                            NegConcPIStart.append(NegConcstartPos.text)
                        for NegConcconceptLen in root5a.iter('Length'):
                            NegConcPILength.append(NegConcconceptLen.text)

    #for testing
    '''
    print(NegCount)
    print(NegType)
    print(NegTrigger)
    print(NegTriggerPIsCount)
    print(NegTriggerPIStart)
    print(NegTriggerPILength)
    print(NegConceptsCount)
    print(NegConceptCUI)
    print(NegConceptMatched)
    print(NegConcPIsCount)
    print(NegConcPIStart)
    print(NegConcPIsCount)                
    '''                


    dict_neg = {
        'NegCount':NegCount,
    'NegType':NegType,
    'NegTrigger':NegTrigger,
    'NegTriggerPIsCount':NegTriggerPIsCount,
    'NegTriggerPIStart':NegTriggerPIStart,
    'NegTriggerPILength':NegTriggerPILength,
    'NegConceptsCount':NegConceptsCount,
    'NegConceptCUI':NegConceptCUI,
    'NegConceptMatched':NegConceptMatched,
    'NegConcPIsCount':NegConcPIsCount,
    'NegConcPIStart':NegConcPIStart,
    'NegTriggerPIStart':NegTriggerPIStart,
    'NegConcPILength':NegConcPILength
    }


    df_neg=pd.DataFrame(dict([(col_name,pd.Series(values)) for col_name,values in dict_neg.items() ]))
    return df_neg
```

```python
def newdf():
    UtterCount=[]
    PMID=[]
    UttSection=[]
    UttNum=[]
    UttText=[]
    UttStartPos=[]
    UttLength=[]
    #Phrases=[]
    PhraseCount=[]
    PhraseText=[]
    SyntaxUnitsCount=[]
    SyntaxType=[]
    LexMatch=[]
    InputMatch=[]
    LexCat=[]
    TokensCount=[]
    Token=[]
    PhraseStartPos=[]
    PhraseLength=[]
    CandidatesTotal=[]
    CandidatesExcluded=[]
    CandidatesPruned=[]
    CandidatesRemaining=[]
    MappingsCount=[]
    MappingScore=[]
    MappingCandidates=[]
    CandidateScore=[]
    CandidateCUI=[]
    CandidateMatched=[]
    CandidatePreferred=[]
    MatchedWordsCount=[]
    MatchedWords=[]
    SemTypesCount=[]
    SemType=[]
    MatchMapsCount=[]
    TextMatchStart=[]
    TextMatchEnd=[]
    ConcMatchStart=[]
    ConcMatchEnd=[]
    LexVariation=[]
    IsHead=[]
    IsOverMatch=[]
    SourcesCount=[]
    Source=[]
    ConceptPIsCount=[]
    StartPos=[]
    Length=[]
    Status=[]
    Negated=[]

    for Utter in root.iter('Utterances'):
        uttroot=et.Element('root')
        uttroot=Utter
        cnt=Utter.attrib['Count']
        if (cnt!='0'):
            UtterCount.append(cnt)
            for utt in uttroot.iter('Utterance'):
                singleUttRoot=et.Element('root')
                singleUttRoot=(utt)
                for pmid in singleUttRoot.iter('PMID'):
                    PMID.append(pmid.text)
                for uttsection in singleUttRoot.iter('UttSection'):
                    UttSection.append(uttsection.text)
                for uttNum in singleUttRoot.iter('UttNum'):
                    UttNum.append(uttNum.text)
                for uttText in singleUttRoot.iter('UttText'):
                    UttText.append(uttText.text)
                for uttStartPos in singleUttRoot.iter('UttStartPos'):
                    UttStartPos.append(uttStartPos.text)
                for uttLength in singleUttRoot.iter('UttLength'):
                    UttLength.append(uttLength.text)
                for Phrases in singleUttRoot.iter('Phrases'):
                    phraseRoot=et.Element('root')
                    phraseRoot=Phrases
                    phraseCnt=Phrases.attrib['Count']
                    if (phraseCnt!='0'):
                        PhraseCount.append(phraseCnt)           

                        for phrase in phraseRoot.iter('Phrase'):
                            singlePhraseRoot=et.Element('root')
                            singlePhraseRoot=phrase
                            for phraseText in singlePhraseRoot.iter('PhraseText'):
                                PhraseText.append(phraseText.text)
                            for SyntaxUnits in singlePhraseRoot.iter('SyntaxUnits'):
                                SyntaxUnitsRoot=et.Element('root')
                                SyntaxUnitsRoot=SyntaxUnits
                                synUnitsCount=SyntaxUnits.attrib['Count']
                                if (synUnitsCount!='0'):
                                    SyntaxUnitsCount.append(synUnitsCount)
                                    for synUnit in SyntaxUnitsRoot.iter('SyntaxUnit'):
                                        singleSynUnitRoot=et.Element('root')
                                        singleSynUnitRoot=synUnit
                                        for syntaxType in singleSynUnitRoot.iter('PhraseText'):
                                            SyntaxType.append(syntaxType.text)
                                        for lexMatch in singleSynUnitRoot.iter('LexMatch'):
                                            LexMatch.append(lexMatch.text)
                                        for inputMatch in singleSynUnitRoot.iter('InputMatch'):
                                            InputMatch.append(inputMatch.text)
                                        for lexCat in singleSynUnitRoot.iter('LexCat'):
                                            LexCat.append(lexCat.text)
                                        for Tokens in singleSynUnitRoot.iter('Tokens'):
                                            TokensRoot=et.Element('root')
                                            TokensRoot=Tokens
                                            TokensCnt=Tokens.attrib['Count']
                                            if (TokensCnt!=0):
                                                TokensCount.append(TokensCnt)
                                                for token in TokensRoot.iter('Token'):
                                                    Token.append(token.text)

                            for phraseStartPos in singlePhraseRoot.iter('PhraseStartPos'): 
                                PhraseStartPos.append(phraseStartPos.text)
                            for phraseLength in singlePhraseRoot.iter('PhraseLength'): 
                                PhraseLength.append(phraseLength.text)
                            for Candidates in singlePhraseRoot.iter('Candidates'): 
                                CandidatesTotal.append(Candidates.attrib['Total'])
                                CandidatesExcluded.append(Candidates.attrib['Excluded'])
                                CandidatesPruned.append(Candidates.attrib['Pruned'])
                                CandidatesRemaining.append(Candidates.attrib['Remaining'])
                            for Mappings in singlePhraseRoot.iter('Mappings'):
                                MappingsRoot=et.Element('root')
                                MappingsRoot=Mappings
                                MappingsCnt=Mappings.attrib['Count']
                                if (MappingsCnt!=0):
                                    MappingsCount.append(MappingsCnt)
                                    for mapping in MappingsRoot.iter('Mapping'):
                                        singleMappingRoot=et.Element('root')
                                        singleMappingRoot=mapping
                                        for mappingScore in singleMappingRoot.iter('MappingScore'):
                                            MappingScore.append(mappingScore.text)
                                        for mappingCandidates in singleMappingRoot.iter('MappingCandidates'):
                                            MapCandidateRoot=et.Element('root')
                                            MapCandidateRoot=mappingCandidates
                                            MappingCandidatesTotal=mappingCandidates.attrib['Total']
                                            if (MappingCandidatesTotal!=0):
                                                MappingCandidates.append(MappingCandidatesTotal)
                                                for Candidate in MapCandidateRoot.iter('Candidate'):
                                                    singleCandidateRoot=et.Element('root')
                                                    singleCandidateRoot=Candidate
                                                    for candidateScore in singleCandidateRoot.iter('CandidateScore'):
                                                        CandidateScore.append(candidateScore.text)
                                                    for candidateCUI in singleCandidateRoot.iter('CandidateCUI'):
                                                        CandidateCUI.append(candidateCUI.text)
                                                    for candidateMatched in singleCandidateRoot.iter('CandidateMatched'):
                                                        CandidateMatched.append(candidateMatched.text)
                                                    for candidatePreferred in singleCandidateRoot.iter('CandidatePreferred'):
                                                        CandidatePreferred.append(candidatePreferred.text)
                                                    for matchedWords in singleCandidateRoot.iter('MatchedWords'):
                                                        MatchedWordsRoot=et.Element('root')
                                                        MatchedWordsRoot=matchedWords
                                                        matchedWordsCount=matchedWords.attrib['Count']
                                                        if (matchedWordsCount!=0):
                                                            MatchedWordsCount.append(matchedWordsCount)
                                                            for matchedWord in MatchedWordsRoot.iter('MatchedWord'):
                                                                MatchedWords.append(matchedWord.text)
                                                    for semTypes in singleCandidateRoot.iter('SemTypes'):
                                                        SemTypesRoot=et.Element('root')
                                                        SemTypesRoot=semTypes
                                                        semTypesCount=semTypes.attrib['Count']
                                                        if (semTypesCount!=0):
                                                            SemTypesCount.append(semTypesCount)
                                                            for semType in SemTypesRoot.iter('SemType'):
                                                                SemType.append(semType.text)
                                                    for matchMaps in singleCandidateRoot.iter('MatchMaps'):
                                                        MatchMapsRoot=et.Element('root')
                                                        MatchMapsRoot=matchMaps
                                                        matchMapsCount=matchMaps.attrib['Count']
                                                        if (matchMapsCount!=0):
                                                            MatchMapsCount.append(matchMapsCount)
                                                            for matchMap in MatchMapsRoot.iter('MatchMap'):
                                                                singleMatchMapRoot=et.Element('root')
                                                                singleMatchMapRoot=matchMap
                                                                for textMatchStart in singleMatchMapRoot.iter('TextMatchStart'):
                                                                    TextMatchStart.append(textMatchStart.text)
                                                                for textMatchEnd in singleMatchMapRoot.iter('TextMatchEnd'):
                                                                    TextMatchEnd.append(textMatchEnd.text)
                                                                for concMatchStart in singleMatchMapRoot.iter('ConcMatchStart'):
                                                                    ConcMatchStart.append(concMatchStart.text)
                                                                for concMatchEnd in singleMatchMapRoot.iter('ConcMatchEnd'):
                                                                    ConcMatchEnd.append(concMatchEnd.text)
                                                                for lexVariation in singleMatchMapRoot.iter('LexVariation'):
                                                                    LexVariation.append(lexVariation.text)
                                                    for isHead in singleCandidateRoot.iter('IsHead'): 
                                                        IsHead.append(isHead.text)
                                                    for isOverMatch in singleCandidateRoot.iter('IsOverMatch'): 
                                                        IsOverMatch.append(isOverMatch.text)
                                                    for sources in singleCandidateRoot.iter('Sources'):
                                                        SourcesRoot=et.Element('root')
                                                        SourcesRoot=sources
                                                        sourcesCount=sources.attrib['Count']
                                                        if (sourcesCount!=0):
                                                            SourcesCount.append(sourcesCount)
                                                            for source in SourcesRoot.iter('Source'):
                                                                Source.append(source.text)                                                      
                                                    for conceptPIs in singleCandidateRoot.iter('ConceptPIs'): 
                                                        ConceptPIsRoot=et.Element('root')
                                                        ConceptPIsRoot=conceptPIs
                                                        conceptPIsCount=conceptPIs.attrib['Count']
                                                        if (conceptPIsCount!=0):
                                                            ConceptPIsCount.append(conceptPIsCount)
                                                            for conceptPI in ConceptPIsRoot.iter('ConceptPI'):
                                                                SingleConceptPIRoot=et.Element('root')
                                                                SingleConceptPIRoot=conceptPI
                                                                for startPos in SingleConceptPIRoot.iter('StartPos'):
                                                                    StartPos.append(startPos.text)
                                                                for length in SingleConceptPIRoot.iter('Length'):
                                                                    Length.append(length.text)

                                                    for status in singleCandidateRoot.iter('Status'): 
                                                        Status.append(status.text)
                                                    for negated in singleCandidateRoot.iter('Negated'): 
                                                        Negated.append(negated.text)


    dict_utt={
    'UtterCount':UtterCount,
    'PMID':PMID,
    'UttSection': UttSection,
    'UttNum':UttNum,
    'UttText': UttText,
    'UttStartPos': UttStartPos,
    'UttLength': UttLength,
    'PhraseCount': PhraseCount,
    'PhraseText': PhraseText,
    'SyntaxUnitsCount': SyntaxUnitsCount,
    'SyntaxType':SyntaxType,
    'LexMatch': LexMatch,
    'InputMatch':InputMatch,
    'LexCat': LexCat,
    'TokensCount': TokensCount,
    'Token': Token,
    'PhraseStartPos':PhraseStartPos,
    'PhraseLength': PhraseLength,
    'CandidatesTotal': CandidatesTotal,
    'CandidatesExcluded': CandidatesExcluded,
    'CandidatesPruned': CandidatesPruned,
    'CandidatesRemaining': CandidatesRemaining,
    'MappingsCount': MappingsCount,
    'MappingScore':MappingScore,    
    'MappingCandidates': MappingCandidates,
    'CandidateScore': CandidateScore,
    'CandidateCUI': CandidateCUI,
    'CandidateMatched': CandidateMatched,
    'CandidatePreferred': CandidatePreferred,
    'MatchedWordsCount': MatchedWordsCount,
    'MatchedWords': MatchedWords,
    'SemTypesCount': SemTypesCount,
    'SemType': SemType,
    'MatchMapsCount': MatchMapsCount,
    'TextMatchStart': TextMatchStart,
    'TextMatchEnd': TextMatchEnd,
    'ConcMatchStart': ConcMatchStart,
    'ConcMatchEnd': ConcMatchEnd,
    'LexVariation': LexVariation,
    'IsHead': IsHead,
    'IsOverMatch': IsOverMatch,
    'SourcesCount': SourcesCount,
    'Source': Source,
    'ConceptPIsCount': ConceptPIsCount,
    'StartPos': StartPos,
    'Length': Length,
    'Status': Status,
    'Negated': Negated
    }


    df=pd.DataFrame(dict([(col_name,pd.Series(values)) for col_name,values in dict_utt.items() ]))
    return df
```

```python
from xml.etree.ElementTree import ParseError
for input_file in list_of_files:
    try:
        tree=et.parse(input_file)
        root=tree.getroot()
        out_df_AA=dfabbreviation()
        out_df_neg=dfnegation()
        out_df=newdf()
        name=input_file.split('.')[0]
        if not os.path.exists(name):
            os.mkdir(name)
        f_AA=name+"_Abbreviation.csv"
        f_Neg=name+"_Negation.csv"
        fname=name+".csv"
        out_df_AA.to_csv(name+'/'+f_AA)
        out_df_neg.to_csv(name+'/'+f_Neg)
        out_df.to_csv(name+'/'+fname)
    except ParseError:
        print("There is an error in parsing "+ input_file)
    
```
